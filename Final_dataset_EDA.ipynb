```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('/content/final_processed_data (1).csv')
```

```python
columns_to_drop = [
    'Unnamed: 0',
    'ID',
    'Severity_Encoded',
    'GMM_Cluster',
    'DBSCAN_Cluster',
    'Cluster',
    'Cluster_k2',
    'Cluster_k3',
    'latitude',
    'longitude',
    'address',
    'address_english',
    'occupation',
    'occupation_english',
    'BCODE',
    'UID1'
]

df = df.drop(columns=columns_to_drop)
```

```python
# Basic information about the dataset
print("Basic Information:")
print(df.info())
```

```python
# Descriptive statistics
print("\nDescriptive Statistics:")
print(df.describe(include='all'))
```

```python
# Unique values per column
print("\nUnique Values per Column:")
for column in df.columns:
    print(f"{column}: {df[column].nunique()} unique values")
```

```python
# Data type counts
print("\nData Type Counts:")
print(df.dtypes.value_counts())
```

```python
# Plotting missing values heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Values Heatmap")
plt.show()
```

```python
# Univariate Analysis
# Plotting distributions of numerical columns
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
for col in numerical_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col].dropna(), kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
# Analyzing categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    plt.figure(figsize=(8, 4))
    sns.countplot(data=df, y=col, order=df[col].value_counts().index)
    plt.title(f"Count Plot of {col}")
    plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

# Encoding all object columns
label_encoder = LabelEncoder()

# Identify object columns
object_columns = df.select_dtypes(include=['object']).columns

# Apply label encoding to each object column
for col in object_columns:
    df[col] = label_encoder.fit_transform(df[col])

# Verify the encoded DataFrame
print("\nEncoded DataFrame:")
print(df.head())
```

```python
# Bivariate Analysis
# Correlation matrix for numerical data

correlation_matrix = df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt='.2f')
plt.title("Correlation Matrix")
plt.show()
```

```python
# Pair plot of numerical columns
sns.pairplot(df[numerical_cols])
plt.show()
```

```python

# Boxplots for categorical vs numerical
for col in categorical_cols:
    for num_col in numerical_cols:
        plt.figure(figsize=(8, 4))
        sns.boxplot(data=df, x=col, y=num_col)
        plt.title(f"Boxplot of {num_col} by {col}")
        plt.xticks(rotation=45)
        plt.show()
```

```python
# Checking for outliers using boxplots
for col in numerical_cols:
    plt.figure(figsize=(8, 4))
    sns.boxplot(data=df, x=col)
    plt.title(f"Boxplot for {col}")
    plt.show()
```

```python
# Analyzing relationships between key variables (e.g., severity and others)
if 'Severity' in df.columns:
    for col in numerical_cols:
        if col != 'Severity':
            plt.figure(figsize=(8, 4))
            sns.boxplot(data=df, x='Severity', y=col)
            plt.title(f"Severity vs {col}")
            plt.show()
```

```python
# List of numerical columns (excluding Severity itself)
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns
numerical_columns = [col for col in numerical_columns if col != 'Severity']
```

```python

# Scatter plots with color-coding based on Severity
for col in numerical_columns:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=df, x=col, y='Severity', hue='Severity', palette='viridis')
    plt.title(f"Scatter Plot: {col} vs Severity (Colored by Severity)")
    plt.xlabel(col)
    plt.ylabel('Severity')
    plt.legend(title="Severity", loc='upper right')
    plt.show()
```
