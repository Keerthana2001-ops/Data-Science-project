```python
import os
import pandas as pd

# Define the folder where your processed CSV files are located
processed_csv_folder = '/content/drive/My Drive/Patient data individual/2024-Data/processed_csv'  # Ensure this path is correct

# List to store results
results = []

# Loop through each CSV file in the folder
for file_name in os.listdir(processed_csv_folder):
    if file_name.endswith('.csv'):
        # Construct the full file path
        file_path = os.path.join(processed_csv_folder, file_name)

        # Read the CSV file
        df = pd.read_csv(file_path)

        # Check if 'C.PEFR' column exists in the DataFrame
        if 'PEFR' in df.columns:
            # Check if 'C.PEFR' values are not integers and convert them if necessary
            if not pd.api.types.is_integer_dtype(df['PEFR']):
                df['PEFR'] = pd.to_numeric(df['PEFR'], errors='coerce')

            # Fill NaN values with the mean of the 'C.PEFR' column
            df['PEFR'].fillna(df['PEFR'].mean(), inplace=True)

            # Convert 'C.PEFR' column to integer type
            df['PEFR'] = df['PEFR'].astype(int)

            # Calculate the average of the 'C.PEFR' column
            average_pefr = df['PEFR'].mean()

            # Append the result with the file name and the average C.PEFR
            results.append({'File Name': str(file_name), 'Average C.PEFR': average_pefr})

            # Optionally, save the updated DataFrame back to the same file if you want to retain the filled values
            # df.to_csv(file_path, index=False)
        else:
            print(f"'C. PEFR' column not found in {file_name}")

# Convert results to a DataFrame and save as a new CSV file
results_df = pd.DataFrame(results)
output_csv_path = '/content/drive/My Drive/Patient data individual/2024-Data/Average_C.PEFR_Summary.csv'
results_df.to_csv(output_csv_path, index=False)
print(f"Summary CSV with averages saved to {output_csv_path}")
```

```python
import os
import pandas as pd
import numpy as np

# Define the folder where your processed CSV files are located
processed_csv_folder = '/content/drive/My Drive/Patient data individual/2024-Data/processed_csv'  # Ensure this path is correct

# List to store results
results = []

# Loop through each CSV file in the folder
for file_name in os.listdir(processed_csv_folder):
    if file_name.endswith('.csv'):
        # Construct the full file path
        file_path = os.path.join(processed_csv_folder, file_name)

        # Read the CSV file
        df = pd.read_csv(file_path)

        # Check if 'C.PEFR' column exists in the DataFrame
        if 'PEFR' in df.columns:
            # Check if 'C.PEFR' values are not integers and convert them if necessary
            if not pd.api.types.is_integer_dtype(df['PEFR']):
                df['PEFR'] = pd.to_numeric(df['PEFR'], errors='coerce')

            # Fill NaN values with the mean of the 'C.PEFR' column
            df['PEFR'].fillna(df['PEFR'].mean(), inplace=True)

            # Convert 'C.PEFR' column to integer type, handling non-finite values
            df['PEFR'] = df['PEFR'].round().astype('Int64')

            # Calculate the average of the 'C.PEFR' column
            average_pefr = df['PEFR'].mean()

            # Append the result with the file name and the average C.PEFR
            results.append({'File Name': str(file_name), 'Average C.PEFR': average_pefr})

            # Optionally, save the updated DataFrame back to the same file if you want to retain the filled values
            # df.to_csv(file_path, index=False)
        else:
            print(f"'C.PEFR' column not found in {file_name}")

# Convert results to a DataFrame and save as a new CSV file
results_df = pd.DataFrame(results)
output_csv_path = '/content/drive/My Drive/Patient data individual/2024-Data/Average_PEFR_Summary.csv'
results_df.to_csv(output_csv_path, index=False)
print(f"Summary CSV with averages saved to {output_csv_path}")
```
